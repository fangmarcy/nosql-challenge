```python
# Define the name of the restaurant you want to search for
restaurant_name = "Penang Flavours"

# Create the query to find the restaurant by its name
query = {"BusinessName": restaurant_name}

# Execute the query to find the restaurant
penang_flavours = establishments.find_one(query)

# Check if the restaurant is found in the database
if penang_flavours:
    # Get the latitude and longitude from the 'geocode' field
    latitude = penang_flavours["geocode"]["latitude"]
    longitude = penang_flavours["geocode"]["longitude"]
    
    print("Penang Flavours Latitude:", latitude)
    print("Penang Flavours Longitude:", longitude)
else:
    print("Penang Flavours not found in the establishments database.")
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
# Define the name of the restaurant you want to search for
restaurant_name = "Penang Flavours"

# Create the query to find the restaurant by its name
query = {"BusinessName": restaurant_name}

# Execute the query to find the restaurant
penang_flavours = establishments.find_one(query)

# Check if the restaurant is found in the database
if penang_flavours:
    # Get the latitude and longitude from the 'geocode' field
    latitude = penang_flavours["geocode"]["latitude"]
    longitude = penang_flavours["geocode"]["longitude"]

    print("Penang Flavours Latitude:", latitude)
    print("Penang Flavours Longitude:", longitude)

    # Define the latitude and longitude range around Penang Flavours (0.01 degrees)
    lat_range = [latitude - 0.01, latitude + 0.01]
    lon_range = [longitude - 0.01, longitude + 0.01]

    # Create the query to find establishments within the latitude and longitude range and with RatingValue = 5
    query_nearby = {
        "RatingValue": 5,  # Limit to establishments with a RatingValue of 5
        "scores.Hygiene": {"$gt": 0},
        "geocode.latitude": {"$gte": lat_range[0], "$lte": lat_range[1]},
        "geocode.longitude": {"$gte": lon_range[0], "$lte": lon_range[1]}
    }

    # Sort the results by hygiene score in ascending order and limit to 5 records
    sort = [("scores.Hygiene", 1)]
    limit = 5

    # Find the establishments matching the query, sorted by hygiene score, and limited to 5 records
    nearby_establishments = establishments.find(query_nearby).sort(sort).limit(limit)

    # Convert the query result (nearby_establishments) to a pandas DataFrame
    df_nearby_establishments = pd.DataFrame(list(nearby_establishments))

    # Display the DataFrame
    print(df_nearby_establishments)
else:
    print("Penang Flavours not found in the establishments database.")
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# Create the aggregation pipeline
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
import pandas as pd
from pprint import pprint

# Create the aggregation pipeline
pipeline = [
    # Match establishments with a hygiene score of 0
    {"$match": {"scores.Hygiene": 0}},
    
    # Group the matched establishments by local authority
    {"$group": {
        "_id": "$LocalAuthorityName",
        "count": {"$sum": 1}
    }},
    
    # Sort the results from highest to lowest based on the count
    {"$sort": {"count": -1}}
]

# Execute the aggregation pipeline
result = establishments.aggregate(pipeline)

# Display the first 10 results using pprint
print("First 10 results:")
for i, doc in enumerate(result):
    pprint(doc)
    if i == 9:  # Print the first 10 results
        break

# Execute the aggregation pipeline again to get the full result
result = establishments.aggregate(pipeline)

# Convert the aggregation result to a pandas DataFrame
df_grouped_and_sorted = pd.DataFrame(list(result))

# Display the number of documents in the result
num_documents = df_grouped_and_sorted.shape[0]
print("Number of documents in the result:", num_documents)

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(df_grouped_and_sorted.head(10))
```
